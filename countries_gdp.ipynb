```python
# Load the file
Countries_gdp = pd.read_csv('Countries_GDP.csv')
```

```python
Countries_gdp
```

```python
# Rename Year columns
Countries_gdp = Countries_gdp.rename(columns={'2000 [YR2000]':'2000',
                             '2001 [YR2001]':'2001',
                             '2002 [YR2002]':'2002',
                             '2003 [YR2003]':'2003',
                             '2004 [YR2004]':'2004',
                             '2005 [YR2005]':'2005',
                             '2006 [YR2006]':'2006',
                             '2007 [YR2007]':'2007',
                             '2008 [YR2008]':'2008',
                             '2009 [YR2009]':'2009',
                             '2010 [YR2010]':'2010',
                             '2011 [YR2011]':'2011',
                             '2012 [YR2012]':'2012',
                             '2013 [YR2013]':'2013',
                             '2014 [YR2014]':'2014',
                             '2015 [YR2015]':'2015',
                             '2016 [YR2016]':'2016',
                             '2017 [YR2017]':'2017',
                             '2018 [YR2018]':'2018',
                             '2019 [YR2019]':'2019'})
```

```python
# Remove any rows with no vlaues
Countries_gdp_clean=Countries_gdp.dropna(how='any')
```

```python
# Remove Unnecessary columns
Countries_gdp_clean = Countries_gdp_clean.drop(columns=['Series Name','Series Code'])
```

```python
Countries_gdp_clean = Countries_gdp_clean.drop(columns=['Country Code'])
```

```python
Countries_gdp_clean
```

```python
# Replace Empty values with space so it allows me to covert all columns into float data type
Countries_gdp_clean = Countries_gdp_clean.replace({'..':''})
```

```python
Countries_gdp_clean
```

```python
columns=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011',
         '2012','2013','2014','2015','2016','2017','2018','2019']
```

```python
# Convert all columns into float so we can run mathmatial calculations
Countries_gdp_clean[columns] = Countries_gdp_clean[columns].apply(pd.to_numeric)
```

```python
Countries_gdp_clean
```

```python
# Calculate average GDP growth % for years 2000 - 2019
gdp_mean = []
years=[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,
         2012,2013,2014,2015,2016,2017,2018,2019]
for i in columns:
    gdp_mean.append(Countries_gdp_clean[str(i)].mean())
```

```python
world_gdp_growth = pd.DataFrame()
world_gdp_growth['Years'] = years
world_gdp_growth['GDP Growth(annual %)'] = gdp_mean
world_gdp_growth
```

```python
plt.style.use('seaborn-ticks')
x_axis=world_gdp_growth['Years']
y_axis=world_gdp_growth['GDP Growth(annual %)']

plt.figure(figsize=(12,8))
plt.plot(x_axis,y_axis,lw=3,
         clip_on=True,marker='o')

for x,y in zip(x_axis,y_axis):
    label = "{:.2f}".format(y)
    plt.annotate(label,# this is the text
                 (x,y),# this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,12), # distance from text to points (x,y)
                 ha='center')
     
plt.grid()
plt.show()
#plt.axis([2000,2019,0,6])
```
